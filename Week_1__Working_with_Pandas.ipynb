### Question 6

Given this sales dataset:

sales = pd.DataFrame({

- Find the total sales per region.

### Question 7

Given a DataFrame with missing values:

data = {'Name': ['Alice', 'Bob', None], 'Score': [90, None, 75]}

### Question 10

You are given a dataset of customer transactions with the following columns:
CustomerID, Product, Category, Quantity, UnitPrice, Date.

Perform the following tasks using pandas:

- Calculate the total purchase amount per transaction (TotalAmount = Quantity * UnitPrice).

- Convert the Date column to a proper datetime format and extract Month and Year.

- Compute monthly total spending per customer.

- Identify the top 2 customers with the highest total spending in each month.

- Create a final DataFrame with columns:

Year, Month, CustomerID, MonthlyTotal, and Rank.

```python


data = {
    'CustomerID': [101, 101, 102, 103, 101, 102, 103, 103],
    'Product': ['Pen', 'Book', 'Pencil', 'Bag', 'Shoes', 'Pencil', 'Pen', 'Book'],
    'Category': ['Stationery', 'Stationery', 'Stationery', 'Bags', 'Footwear', 'Stationery', 'Stationery', 'Stationery'],
    'Quantity': [2, 1, 3, 1, 2, 4, 5, 2],
    'UnitPrice': [1.5, 5.0, 0.5, 20.0, 50.0, 0.5, 1.5, 5.0],
    'Date': ['2024-01-05', '2024-01-10', '2024-01-15', '2024-01-15', '2024-02-01', '2024-02-05', '2024-02-07', '2024-02-15']
}


```

```python
# convert the data into a DataFrame
df = pd.DataFrame(data)

# show df
df
```

```python
# do some EDA
# summary statistics of the numnerical data
df.describe()
```

```python
# checking for null values
df.isnull().sum()
```

```python
# step 1: compute the total purchase amount
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
df
```

```python
# step 2: Convert Date and extract Month and year
df['Date'] = pd.to_datetime(df['Date'])

# extracting the month
df['Month'] = df['Date'].dt.month

# extrating the year
df['Year'] = df['Date'].dt.year

df
```

```python
# step 3: Computing the monthly total spending per customer
monthly = df.groupby(['Year', 'Month', 'CustomerID'])['TotalAmount'].sum().reset_index(name = 'MonthlyTotal')
```

```python
df
```

```python
# step 4: Rank customers per month by spending
monthly['Rank'] = monthly.groupby(['Year', 'Month'])['MonthlyTotal'].rank(method = 'dense', ascending = 'False')
```

```python
# step 5: Filter by top 2 customers by month
top_customers = monthly[monthly['Rank'] <= 2].sort_values(['Year', 'Month', 'Rank'])
top_customers
```
